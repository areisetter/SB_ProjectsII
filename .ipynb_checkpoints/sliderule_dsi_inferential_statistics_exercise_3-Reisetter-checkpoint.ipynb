# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
import scipy.stats as stat
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

****
## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

A.  It's a decent first glance at the data, but there's not enough information to make these recommendations.  Scatter plots like these draw attention to outliers and do not convey density of points, and therefore do not lend themselves to valuable statistical conclusions (unless the points are considerably more sparse).  The statistical statements above, while true, are not complete and may be misleading (I need to look into it more).  The high discharge hospitals are all outliers and should not guide the conclusions.  The recommendations based on hospital size look to be very strange -- the ratio still looks like a much better test of hospital performance than the size.  We should do correlation testing.  Ideally, then we'd be able to look at details for the hospitals and the nature of the discharges (hospitals, especially small ones, may specialize in diseases or populations that tend to result in readmission).  The write-up of the current HRRP program says that these demographics are taken into account in the dataset, though, so I won't worry about those for now.

```python
x = clean_hospital_read_df['Number of Discharges']
y = clean_hospital_read_df['Excess Readmission Ratio']

#documentation states that hospitals need >=25 discharges to be evaluated
#I don't want the outliers at high discharge to pull lin regress. around
cutx=x[x>=25][x<1000]
cuty=y[x>=25][x<1000]

#as a first try, I'll do linear regression to see what it gets me
#first do density of points plot so I can see what's going on
plt.hist2d(cutx, cuty, bins=(50, 50), cmap=plt.cm.Purples)
plt.show()
#a linear regress will find fit AND test slope=0 hypothesis
stat.linregress(cutx,cuty)

```

The linear regression was a lousy fit (rvalue close to 0), indicating that a conclusion stating a correlation or linear dependence of ratio on size of hospital would not be justified.  But the p-value for the 0-slope fit was also small, so simply drawing a flat line through the data is not a good model.
I'm going to try to bin the data (low discharges vs high discharges) and check to see if the average readmission ratio is statistically significant in the two bins.  The null hypothesis is that they're statistically equivalent.  If it's consistent with the linregress test, the result will not be consistent with the null hypothesis.

```python
lowx=x[x>25][x<500]
lowy=y[x>25][x<500]

highx=x[x>1000]
highy=y[x>1000]

stat.ttest_ind(lowy,highy)
```

It's not looking good for the null hypothesis.  It looks like there's a 7-sigma difference in the low- and high- discharge rate groups.
Just for the record, I tried several different binning techniques, including smaller bins with different means, and the significance didn't go away.

Question 5.  It might be more helpful to look at the fraction of hospitals above a ratio of 1.0 in small bins of discharge rates.  This will more directly present what the original analysis was getting at.  Alternatively, the point density (2d histogram) figure above also gives similar information.  I'm going to make the histogram below.

```python
binwidth=20
minx=25 
maxx=1600
nbins=int(maxx/binwidth)
centers=list()
fractions=list()
bars=list()
for i in range(1,nbins):
        middle=binwidth*i-binwidth/2+minx
        centers.append(middle)
        numerator=y[y>=1][x>=middle-binwidth/2][x<middle+binwidth/2].count()
        denominator=y[x>=middle-binwidth/2][x<middle+binwidth/2].count()
        fractions.append(numerator/denominator)
        bars.append(denominator**(1/2)/denominator*numerator/denominator)
fig, ax = plt.subplots(figsize=(8,4))
ax.errorbar(centers, fractions,yerr=bars, fmt='o')    
ax.grid(axis='y')
ax.set_xlabel('Number of discharges', fontsize=11)
ax.set_ylabel('Fraction of hospitals with  readmission ratio >1', fontsize=11)
ax.set_title('Fraction of hospitals with high readmission rates vs. hospital size (discharges)')


            
```
